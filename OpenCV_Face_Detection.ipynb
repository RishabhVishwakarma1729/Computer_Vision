```python
import cv2
import matplotlib.pyplot as plt
```

```python
arphoto=cv2.imread('C:/Users/Asus/Desktop/Antrix/cricket1.jpg', 0)
arphoto.shape
```

```python
face_cascade=cv2.CascadeClassifier("C:/Users/Asus/Desktop/Antrix/haarcascade_frontalface_default.xml")
```

```python
faces=face_cascade.detectMultiScale(arphoto,
                                    scaleFactor=1.25,
                                    minNeighbors=5)
```

```python
faces
```

```python
for x,y,w,h in faces:   
    img=cv2.rectangle(arphoto, (x,y), (x+w,y+h),(0,255,0),3)
```

```python
plt.imshow(img)
```

```python
faces = [
    [147,  86,  27,  27],
    [311,  90,  28,  28],
    [384,  92,  29,  29],
    [118,  28,  27,  27],
    [184,  26,  27,  27],
    [373,  34,  28,  28],
    [307,  35,  27,  27],
    [ 48,  43,  27,  27]
]    
```

```python
# Initialize list to store the detected face images
detected_faces = []

for i, (x, y, w, h) in enumerate(faces):
    # Slice out the face from the image using the provided coordinates
    face = arphoto[y:y+h, x:x+w]
    detected_faces.append(face)
    
# Display each detected face
    plt.subplot(1, len(faces), i + 1)
    plt.imshow(face, cmap='gray')
    plt.axis('off')
    plt.title(f"Face {i + 1}")

plt.show()
```
